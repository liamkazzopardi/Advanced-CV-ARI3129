```python
from ultralytics import YOLO

model = YOLO('yolov12n.pt')

model.train(
    data='/content/data.yaml',
    epochs=100,
    imgsz=640,
    batch=-1,
    device=0,

    # Training control
    patience=20,

    # Small object detection tweaks
    mosaic=1.0,         # Keep mosaic augmentation (good for small objects)
    close_mosaic=10,    # But disable at end for precision

    # Augmentation for signs (which can appear at different scales/angles)
    degrees=15,         # Rotation augmentation (signs at angles)
    scale=0.5,          # Scale augmentation
    fliplr=0.0,         # DON'T flip horizontally - signs have specific orientations!

    # Optimization
    optimizer='auto',

    # Organization
    project='runs/detect',
    name='traffic_signs_yolov12n2',
    exist_ok=True
)
```
