### Verify Dataset

```python
import os

dataset_path = "/Users/matthewprivitera/Documents/School/University/3rd Year/Semester 1/Advanced Computer Vision for AI/Assignment/assignment_materials/YOLO_COCO_view_angle"
yaml_path = "/Users/matthewprivitera/Documents/School/University/3rd Year/Semester 1/Advanced Computer Vision for AI/Assignment/assignment_materials/YOLO_COCO_view_angle/data.yaml"

print("\nDataset verification:")
for split in ['train', 'val', 'test']:
    img_count = len(os.listdir(os.path.join(dataset_path, 'images', split)))
    lbl_count = len(os.listdir(os.path.join(dataset_path, 'labels', split)))
    print(f"  {split}: {img_count} images, {lbl_count} labels")
```

### Train YOLOv12 Nano Model

```python
from ultralytics import YOLO

# Load pretrained model
model = YOLO("yolo12n.pt")

results = model.train(
    data=yaml_path, 
    epochs=100, 
    imgsz=640, 
    batch=32,
    device='mps',
    workers = 4,
    patience = 20,
    project = 'runs/viewing_angle',
    name='Maltese_Signs_ViewingAngle',
    optimizer = "AdamW",
    lr0=0.001,
    save_period = 10,
    augment=True,
    plots=True,
    verbose=True
)
```

```python
from ultralytics import YOLO

model = YOLO('best.pt')

print("Complete Evaluation - All Splits")

# Train
print("\n1. Evaluating TRAIN set (488 images)...")
train = model.val(data=yaml_path, split='train')
print(f"Train mAP@50-95: {train.box.map:.4f}")
print(f"Images evaluated: Check output above ‚Üë")

# Val
print("\n2. Evaluating VAL set (93 images)...")
val = model.val(data=yaml_path, split='val')
print(f"Val mAP@50-95: {val.box.map:.4f}")

# Test
print("\n3. Evaluating TEST set (122 images)...")
test = model.val(data=yaml_path, split='test')
print(f"Test mAP@50-95: {test.box.map:.4f}")

print("Overfitting Check")
print(f"Train: {train.box.map:.4f} ({train.box.map*100:.1f}%)")
print(f"Val:   {val.box.map:.4f} ({val.box.map*100:.1f}%)")
print(f"Test:  {test.box.map:.4f} ({test.box.map*100:.1f}%)")
print(f"\nTrain‚ÜíTest gap: {(train.box.map - test.box.map):.4f} ({(train.box.map - test.box.map)/train.box.map*100:.1f}%)")
```
